```python
kmn= KMeans(n_clusters=2)
kmn.fit(iris.data)
labels = kmn.predict(iris.data)
```

```python
centroids= kmn.cluster_centers_
```

```python
kmn.inertia_
```

```python
inertia_list = []

for k in np.arange(1, 6):
  kmn = KMeans(n_clusters= k)
  kmn.fit(iris.data)
  inertia_list.append(kmn.inertia_)
inertia_list
```

```python
plt.plot(np.arange(1, 6), inertia_list, 'bo-')
plt.xlabel('number of cluster')
plt.ylabel('inertia')
plt.show()
```

```python
plt.plot(np.arange(1, 6), inertia_list, 'ro-')
plt.xlabel('number of cluster')
plt.ylabel('inertia')
plt.show()
```



# Hierarchical clustering 
    
                 





```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
hierachical= linkage(iris.data, method='complete')
dendrogram(hierachical)
plt.show()
```

```python
labels= fcluster(hierachical, 3, criterion='distance')
labels
```

```python
plt.scatter(iris.data[: ,0], iris.data[: ,2], c=labels)
plt.show()
```

```python
labels= fcluster(hierachical, 6, criterion='distance')
plt.scatter(iris.data[: ,0], iris.data[: ,2], c=labels)
plt.show()

```

```python
labels= fcluster(hierachical,2, criterion='distance')
plt.scatter(iris.data[: ,0], iris.data[: ,2], c=labels)
plt.show()
```

#Meanshift
  kmeans        cluster   
 
  bandwidth


```python
from sklearn.cluster import MeanShift
```

```python
ms= MeanShift()
ms.fit(iris.data)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
plt.scatter(iris.data[ : ,0], iris.data[ : , 2], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[ :, 2], marker='x', s=150, linewidth=5)
plt.show()
```

```python
ms= MeanShift(bandwidth=5)
ms.fit(iris.data)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
plt.scatter(iris.data[ : ,0], iris.data[ : , 2], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[ :, 2], marker='x', s=150, linewidth=5)
plt.show()
```

#DBSCAN 
Density-Based spatial clustering of Applications with Noise

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan= DBSCAN()
dbscan.fit(iris.data)
labels= dbscan.labels_
```

```python
plt.scatter(iris.data[ : , 0], iris.data[ : , 2], c=labels)
plt.show()
```

```python
plt.scatter(iris.data[ : , 1], iris.data[ : , 2], c=labels)
plt.show()
```

```python
dbscan= DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(iris.data)
labels= dbscan.labels_
labels
```

```python
dbscan= DBSCAN(eps=0.1, min_samples=5)
dbscan.fit(iris.data)
labels= dbscan.labels_
labels
```
